```python
# Importando biblioteca
import os
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# This all the files available to analysis about the ecommerce sales
os.listdir('/content/brazilian-ecommerce')
```

```python
# Reading all the files
raw_path = './brazilian-ecommerce/'
olist_customer = pd.read_csv(raw_path + 'olist_customers_dataset.csv')
olist_geolocation = pd.read_csv(raw_path + 'olist_geolocation_dataset.csv')
olist_orders = pd.read_csv(raw_path + 'olist_orders_dataset.csv')
olist_order_items = pd.read_csv(raw_path + 'olist_order_items_dataset.csv')
olist_order_payments = pd.read_csv(raw_path + 'olist_order_payments_dataset.csv')
olist_order_reviews = pd.read_csv(raw_path + 'olist_order_reviews_dataset.csv')
olist_products = pd.read_csv(raw_path + 'olist_products_dataset.csv')
olist_sellers = pd.read_csv(raw_path + 'olist_sellers_dataset.csv')
```

### Dataset olist_orders

```python
# Check if exist null values
olist_orders.isnull().sum()
```

```python
 # Replace the missing values with the function 'ffill' stands for 'forward fill' and will propagate last valid observation forward. inplace
olist_orders['order_delivered_customer_date'] = olist_orders['order_delivered_customer_date'].fillna(method='ffill')
olist_orders['order_delivered_carrier_date'] = olist_orders['order_delivered_carrier_date'].fillna(method='ffill')
olist_orders['order_approved_at'] = olist_orders['order_approved_at'].fillna(method='ffill')
olist_orders.isnull().sum()
```

```python
# Changing the data type
olist_orders.order_purchase_timestamp = pd.to_datetime(olist_orders.order_purchase_timestamp)
```

```python
print("Total number of orders in the database:",olist_orders['order_id'].nunique())
print("Total Number of customers:",olist_orders['customer_id'].nunique())
```

```python
# Lets check the order status
status = olist_orders.groupby('order_status')['order_id'].nunique().sort_values(ascending=False)
status
```

```python
# Orders.dtypes
orders_mod = olist_orders.copy()
orders_mod["order_purchase_timestamp"] = pd.to_datetime(olist_orders["order_purchase_timestamp"], format='%Y-%m-%d %H:%M:%S')
orders_mod["order_delivered_carrier_date"] = pd.to_datetime(olist_orders["order_delivered_carrier_date"], format='%Y-%m-%d %H:%M:%S')
orders_mod["order_delivered_customer_date"] = pd.to_datetime(olist_orders["order_delivered_customer_date"], format='%Y-%m-%d %H:%M:%S')
orders_mod["order_estimated_delivery_date"] = pd.to_datetime(olist_orders["order_estimated_delivery_date"], format='%Y-%m-%d %H:%M:%S')
orders_mod.dtypes
```

```python
counts = orders_mod.set_index("order_purchase_timestamp").groupby(pd.Grouper(freq='D')).count()
fig = plt.figure(figsize=(16,8))
ax = fig.gca()
counts.plot(y = "order_id", use_index=True, ax=ax)
```

In the graph you can see that in Christmas the sales increase.

#### Analysis of the Customers

```python
cities = olist_customer["customer_city"].nunique()
c1 = olist_customer.groupby('customer_city')['customer_id'].nunique().sort_values(ascending=False)
print("There are",cities,"unique cities in the database. The TOP 10 cities are:")
c2 = c1.head(10)
print(c2)
print("\nTOP 10 cities covers", round(c2.sum()/olist_customer.shape[0]*100,1),"percent of all the orders.")
plt.figure(figsize=(16,8))
c2.plot(kind="bar",rot=0)
```

#### Analysis of the Payments 

```python
print(olist_order_payments.describe())
print(olist_order_payments["payment_type"].value_counts())
credit = olist_order_payments[olist_order_payments["payment_type"]=="credit_card"]
```

The most common method of payment is with credit card.

```python
fig, ax = plt.subplots(figsize=(16,8))
credit.hist(column = "payment_installments", bins = credit["payment_installments"].max(), ax=ax)
plt.xlabel("Number of installments")
plt.ylabel("Counts")
plt.title("Histogram of installments count")
```

In thisthis graph we can say for sure is credit card with 74%

```python
fig, ax = plt.subplots(figsize=(9, 8), subplot_kw=dict(aspect="equal"))
explode = (0.1, 0, 0, 0)
colors = ['#f45a5a', '#449dfc', '#93f96d', '#f9c86d']
legend = ["Credit Card", "Boleto", "Voucher", "Debit Card"]

p = olist_order_payments["payment_type"][olist_order_payments["payment_type"] != "not_defined"].value_counts()
p.plot(kind="pie", legend=False, labels=None, startangle=0, explode=explode, autopct='%1.0f%%', pctdistance=0.6, shadow=True, textprops={'weight':'bold', 'fontsize':16}, 
       colors=colors, ax=ax)
ax.legend(legend, loc='best', shadow=True, prop={'weight':'bold', 'size':12}, bbox_to_anchor=(0.6, 0, 0.5,1))
plt.title("Paymement methods", fontweight='bold', size=16)
plt.ylabel("")
```

Based in the category we can say cama_mesa_banho was the most popular category in the dataset

```python
import plotly.graph_objects as go

fig = go.Figure(go.Bar(
    x=popular_categories.head(10).index,y=popular_categories['order_id'].head(10),
    marker={'color': popular_categories['order_id'], 
    'colorscale': 'Viridis'},  
    text=popular_categories['order_id'],
    textposition = "outside",
))
fig.update_layout(title_text='Most popular categories',xaxis_title="Category Name",yaxis_title="No. of orders")
fig.show()
```
